```python
def calc_total_bill(price, tax_rate=0.13, tip_rate="standard"):
    tip_presets = {
        "low": 0.10,
        "standard": 0.15,
        "high": 0.20
    }

    # only convert if it's one of the known presets
    if tip_rate in tip_presets:
        tip_rate = tip_presets[tip_rate]

    tax = price * tax_rate
    tip = price * tip_rate
    total = price + tax + tip
    return total, tax, tip
```

```python
calc_total_bill(100, tip_rate="low")
```

```python
def eye_exam_covered(age, qualifying_condition, time_since_last_exam):
    # assumptions:
    # age is int
    # qualifying_condition is bool, True if you have eligable condition, otherwise False
    # time_since_last_exam is int (in months)
    if age <= 19:
        if time_since_last_exam >= 12:
            return "You are eligable for one major eye exam and any minor assessments needed."
        else:
            return "It has not been 12 months since your last exam, /n but you can have a minior assessment if needed."
    elif 20 <= age <= 64:
        if qualifying_condition:
            if time_since_last_exam >= 12:
                return "you are eligable for 1 majot eye exam and 2 additional follow-up minor assessments."
            else:
                return "It has not been 12 months since your last exam, /n but you can have a minior assessment if needed."
        else:
            return "You do not have an sligable medical condition"
    elif age >= 65:
        if time_since_last_exam >= 12:
            return "You are eligable for one major eye exam and 2 follow up minor assessments needed."
        else:
            return "It has not been 12 months since your last exam, /n but you can have a minior assessment if needed."
    else:
        return "Invalid age input."
            
```
